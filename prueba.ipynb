```python
from reportlab.lib.pagesizes import A4
from reportlab.pdfgen import canvas
from reportlab.lib import colors
from reportlab.lib.units import cm

def generate_routine_pdf(routine_data, filename="rutina.pdf"):
    """
    Genera un PDF con m칰ltiples rutinas, separando cada una en una p치gina nueva.
    """
    c = canvas.Canvas(filename, pagesize=A4)
    width, height = A4
    margin = 2 * cm

    for idx, routine in enumerate(routine_data):
        # Nueva p치gina para cada rutina
        if idx > 0:
            c.showPage()

        c.setFont("Helvetica-Bold", 16)
        c.drawString(margin, height - margin, f"Rutina {idx + 1}")
        c.setStrokeColor(colors.black)
        c.line(margin, height - margin - 5, width - margin, height - margin - 5)

        y_position = height - margin - 30

        for i, exercise in enumerate(routine["rutina"]):
            if y_position < 50:  # Salto de p치gina si no hay espacio
                c.showPage()
                c.setFont("Helvetica-Bold", 16)
                c.drawString(margin, height - margin, f"Rutina {idx + 1} (Cont.)")
                c.line(margin, height - margin - 5, width - margin, height - margin - 5)
                y_position = height - margin - 30

            # Escribir el nombre del ejercicio
            c.setFont("Helvetica-Bold", 12)
            c.drawString(margin, y_position, f"{i+1}. {exercise['ejercicio']}")

            y_position -= 15

            # Escribir series y repeticiones
            c.setFont("Helvetica", 11)
            c.drawString(margin + 20, y_position, f"Series: {exercise['series']}")
            y_position -= 15

            repeticiones = ", ".join(exercise["repeticiones"])
            c.drawString(margin + 20, y_position, f"Repeticiones: {repeticiones}")
            y_position -= 30  # Espaciado entre ejercicios

    c.save()
    return filename  # Retorna el nombre del archivo PDF generado

```

```python

def generate_html_presentation(routine_data, filename="rutina.html"):
    """
    Genera una presentaci칩n HTML bien estructurada con CSS.
    """
    html_content = """
    <!DOCTYPE html>
    <html lang="es">
    <head>
        <meta charset="UTF-8">
        <meta name="viewport" content="width=device-width, initial-scale=1.0">
        <title>Rutina de Entrenamiento</title>
        <style>
            body {
                font-family: Arial, sans-serif;
                background-color: #f5f5f5;
                color: #333;
                text-align: center;
                margin: 0;
                padding: 0;
            }
            .slide {
                width: 80%;
                margin: 50px auto;
                background: white;
                padding: 20px;
                border-radius: 10px;
                box-shadow: 2px 2px 10px rgba(0,0,0,0.1);
                display: none;
            }
            h1 {
                color: #0056b3;
            }
            table {
                width: 100%;
                border-collapse: collapse;
                margin-top: 10px;
            }
            th, td {
                border: 1px solid #ddd;
                padding: 10px;
                text-align: left;
            }
            th {
                background-color: #0056b3;
                color: white;
            }
            .controls {
                margin-top: 20px;
            }
            button {
                padding: 10px 20px;
                font-size: 16px;
                border: none;
                cursor: pointer;
                margin: 5px;
            }
            .prev { background-color: #666; color: white; }
            .next { background-color: #0056b3; color: white; }
        </style>
    </head>
    <body>
        <h1>Rutina de Entrenamiento</h1>
    """

    # Generar cada rutina como una diapositiva
    for idx, routine in enumerate(routine_data):
        html_content += f"""
        <div class="slide" id="slide-{idx}">
            <h2>Rutina {idx + 1}</h2>
            <table>
                <tr>
                    <th>Ejercicio</th>
                    <th>Series</th>
                    <th>Repeticiones</th>
                </tr>
        """

        for exercise in routine["rutina"]:
            repeticiones = ", ".join(exercise["repeticiones"])
            html_content += f"""
            <tr>
                <td>{exercise['ejercicio']}</td>
                <td>{exercise['series']}</td>
                <td>{repeticiones}</td>
            </tr>
            """

        html_content += "</table></div>"

    # Agregar controles de navegaci칩n
    html_content += """
    <div class="controls">
        <button class="prev" onclick="changeSlide(-1)">Anterior</button>
        <button class="next" onclick="changeSlide(1)">Siguiente</button>
    </div>

    <script>
        let currentSlide = 0;
        const slides = document.querySelectorAll(".slide");

        function showSlide(index) {
            slides.forEach(slide => slide.style.display = "none");
            slides[index].style.display = "block";
        }

        function changeSlide(direction) {
            currentSlide += direction;
            if (currentSlide >= slides.length) currentSlide = 0;
            if (currentSlide < 0) currentSlide = slides.length - 1;
            showSlide(currentSlide);
        }

        showSlide(currentSlide);
    </script>
    </body>
    </html>
    """

    # Guardar el HTML
    with open(filename, "w", encoding="utf-8") as f:
        f.write(html_content)

    return filename
```

```python
import time
from googleapiclient.discovery import build
from google.oauth2 import service_account

# 游댳 Cargar credenciales
SCOPES = ["https://www.googleapis.com/auth/forms.body", "https://www.googleapis.com/auth/drive"]
SERVICE_ACCOUNT_FILE = "credenciales.json"  # Aseg칰rate de que este archivo es correcto

credentials = service_account.Credentials.from_service_account_file(
    SERVICE_ACCOUNT_FILE, scopes=SCOPES
)
forms_service = build("forms", "v1", credentials=credentials)

def create_google_form(routine_data):
    """
    Crea un Google Form con los ejercicios de la rutina y evita valores duplicados en las respuestas.
    """
    # 游댳 Crear el formulario
    form_metadata = {
        "info": {
            "title": "Formulario de Rutina de Entrenamiento",
            "documentTitle": "Rutina de Entrenamiento Personalizada"
        }
    }
    
    form = forms_service.forms().create(body=form_metadata).execute()
    form_id = form["formId"]
    
    # 游댳 Agregar preguntas al formulario
    requests = []
    
    for i, rutina in enumerate(routine_data):
        for exercise in rutina["rutina"]:
            # 游댳 Eliminar valores duplicados en repeticiones
            unique_reps = list(set(exercise["repeticiones"]))  # Convertir a conjunto y luego a lista para quitar duplicados
            
            # 游댳 Verificar que haya opciones v치lidas
            if not unique_reps:
                unique_reps = ["N/A"]  # Agregar una opci칩n si la lista est치 vac칤a
            
            requests.append({
                "createItem": {
                    "item": {
                        "title": f"{exercise['ejercicio']} - {exercise['series']} series",
                        "questionItem": {
                            "question": {
                                "required": True,
                                "choiceQuestion": {
                                    "type": "RADIO",
                                    "options": [{"value": rep} for rep in unique_reps],
                                    "shuffle": False
                                }
                            }
                        }
                    },
                    "location": {"index": i}
                }
            })
    
    # 游댳 Enviar las preguntas al formulario
    forms_service.forms().batchUpdate(formId=form_id, body={"requests": requests}).execute()
    
    # 游댳 Hacer el formulario accesible
    set_permissions(form_id)
    
    # 游댳 Devolver el enlace del formulario
    return f"https://docs.google.com/forms/d/{form_id}/edit"

def set_permissions(file_id):
    """
    Da permisos de acceso p칰blico al formulario.
    """
    drive_service = build("drive", "v3", credentials=credentials)
    
    permission = {
        "type": "anyone",
        "role": "writer"  # Permite que cualquiera lo complete
    }

    drive_service.permissions().create(
        fileId=file_id,
        body=permission
    ).execute()

```
